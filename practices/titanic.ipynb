```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 读取数据
file_path = 'resources/data/titanic/'
train_data = pd.read_csv(file_path + 'train.csv')

train_data.head()
```

```python
test_data = pd.read_csv(file_path + 'test.csv')
test_data.head()
```

```python
from sklearn.ensemble import RandomForestClassifier

features = ["Pclass", "Sex", "SibSp", "Parch"]
y_train_data = train_data['Survived']
X = pd.get_dummies(train_data[features]) # 获取one-hot编码
X_test = pd.get_dummies(test_data[features])
```

```python
y_train_data
```

```python
X
```

```python
X_test
```

```python
model = RandomForestClassifier(random_state=1)
model.fit(X, y_train_data)
predictions = model.predict(X_test)
```

```python
predictions
```

```python
output = pd.DataFrame(
    {'PassengerId': test_data.PassengerId,
     'Survived': predictions
    }
)
```

```python
output
```

```python
output.to_csv('my_submisson.csv', index=False)
print('successfully!')
```

```python
from sklearn.datasets import load_iris 
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
iris = load_iris()
```

```python
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3)
```

```python
train_data = pd.concat([pd.DataFrame(x_train), pd.DataFrame(y_train)], axis=1)
```

```python
kg = KNeighborsClassifier(n_neighbors=15, algorithm="kd_tree")
```

```python
kg.fit(x_train, y_train)
```

```python
kg.score(x_test, y_test)
```

```python
score = []
for i in range (3, 50):
    kg = KNeighborsClassifier(n_neighbors=i)
    kg.fit(x_train, y_train)
    score.append(kg.score(x_test, y_test))
plt.plot(range(3, 50), score, label='max_n')
plt.legend()
plt.show()
```
